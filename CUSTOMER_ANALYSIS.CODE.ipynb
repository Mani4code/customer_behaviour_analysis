```python
%%sql
##GENERATING INSIGHTS FROM THE DATA 
#Q1--> what is the total revenue generated bu male vs female custemers?
select gender,sum(purchase_amount) as total_purchase
from coustmer_table 
group by gender
```

```python
%%sql
#Q2--->Which top 5 product has highest review_rating among all the products
SELECT item_purchased, ROUND(AVG(review_rating), 2) AS average_rating
FROM coustmer_table
GROUP BY item_purchased
ORDER BY AVG(review_rating) DESC
LIMIT 5;
```

```python
%%sql
#Q3--->compare the average amount of purchase between standard and express shipping

select shipping_type, 
       ROUND(AVG(purchase_amount), 2) AS avg_purchase_amount
from coustmer_table
where shipping_type IN ('Standard', 'Express')
group by shipping_type;
```

```python
%%sql
#Q4-->Q5. DO subscribed customers spend more? Compare average spend and total revenue between subscribers and mon-subscribers.
select subscription_status,
COUNT(customer_id) AS TOTAL_CUSTOMERS,
ROUND(AVG(purchase_amount),2) AS AVG_SPENDING,
ROUND(SUM(purchase_amount),2) AS TOTAL_REVENUE
FROM coustmer_table 
GROUP BY subscription_status
ORDER BY AVG_SPENDING,TOTAL_REVENUE DESC;
```

```python
%%sql
##Q5-->Which 5 products has the highest percentage of purchases with discounts applied

select item_purchased,count(item_purchased) as PURCHASE_COUNT,
ROUND(100 * SUM(CASE WHEN discount_applied = 'yes' THEN 1 ELSE 0 END)/COUNT(*),2) AS DISCOUNT_RATE
from coustmer_table 
group by  item_purchased
order by DISCOUNT_RATE desc
limit 5;
```

```python
%%sql
#Q6-->Segment customers  as NEW,RETURNING AND LOYAL based on Their prviouspurchases and  list their no.of purchases
with customers_type as (
select customer_id,previous_purchases ,
    CASE
    WHEN previous_purchases = 1 THEN 'NEW'
    WHEN  previous_purchases BETWEEN 2 AND 10 THEN 'RETURNING'
    ELSE 'LOYAL'
END AS CUSTOMER_SEGMENT
FROM coustmer_table) 

SELECT  CUSTOMER_SEGMENT, COUNT(*) AS "NO.OF CUSTOMERS"
FROM customers_type 
GROUP BY CUSTOMER_SEGMENT
```

```python
%%sql
##Q7-->which is the top 3 puruchased items of each category
with items_count as(
select category,
item_purchased,
count(customer_id) as total_orders,
ROW_NUMBER() over(partition by category order by count(customer_id) desc) as item_rank
FROM coustmer_table
group by category, item_purchased
)

select item_rank ,category, item_purchased, total_orders 
from items_count
where item_rank <= 3;
```

```python
%%sql
##Q8--->Are the customers who are repeat buyers (more than 5 previous purchases) also likely to subscribe?
select subscription_status,
count(customer_id) as repeat_buyers
from coustmer_table
where  previous_purchases > 5
group by subscription_status  
order by repeat_buyers;
```

```python
%%sql
##Q9--->What is the revenue contribution of each age group

select age_group,
sum(purchase_amount) as total_purschase_amount
from coustmer_table
group by age_group
order by total_purschase_amount desc;
```
